```python
import numpy as np
import pygmt
```

```python
xs = np.arange(-180, 180, 60)
ys = np.arange(-90, 90, 30)
longs, lats = np.meshgrid(xs, ys)
```

```python
grid = pygmt.datasets.load_earth_relief(resolution='01d')
```

```python
fig = pygmt.Figure()

fig.grdimage(grid=grid,
             projection='Q0/0/12c',
             frame='af',
             cmap='oleron')

fig.coast(resolution='l', shorelines='0.005c', area_thresh=5000)

fig.plot(x=longs.flatten(),
         y=lats.flatten(),
         style='E0/1000/1000',
         fill='darkorange',
         pen='black',
         transparency=60)

fig.show()
```
